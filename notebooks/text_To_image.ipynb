```python
# diffusers-->hugging face page to use diffusion models from huffing face page
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt
import gradio as gr
```

```python
!pip show torch
```

```python
# can use any
model_id1="dreamlike-art/dreamlike-diffusion-1.0"
model_id2="stabilityai/stable-diffusion-xl-base-1.0"
```

```python
pipe=StableDiffusionPipeline.from_pretrained(model_id1,torch_dtype=torch.float16,use_auth_token=True,USE_SAFETENSORS=True)
pipe.to("cuda")
```

```python
promt = "a photo of an astronaut riding a horse on the moon"
```

```python
image=pipe(promt).images[0]
```
